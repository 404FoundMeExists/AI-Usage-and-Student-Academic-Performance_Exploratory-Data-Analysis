OVERVIEW DATAFRAME

1. age : usia siswa dalam tahun

2. jenjang_pendidikan : tingkat pendidikan siswa saat ini (sekolah, perguruan tinggi).

3. rerata_jam_perhari : jumlah rata rata jam belajar siswa per hari.

4. penggunaan : menunjukkan apakah siswa menggunakan alat AI untuk tujuan belajar

5. tools_ai : jenia alat AI yang digunakan oleh siswa

6. tujuan : indikator kinerja akademik seperti nilai atau tingkat kinerja.

7. nilai_sebelum : nilai akademik siswa sebelum menggunakan alat AI.

8. nilai_sesudah : nilai akademik sesudah menggunakan alat AI.

9. rerata_waktu_perhari : rata-rata harian siswa dalam jam

```python
df_ai.groupby('tujuan')['delta_nilai'].agg(['count', 'mean', 'std'])
```

Analisis deskriptif berdasarkan tujuan penggunaan AI menunjukkan bahwa rata-rata kenaikan nilai relatif berbeda antar tujuan. Tujuan penggunaan AI untuk kegiatan research dan coding memiliki rata-rata peningkatan nilai yang sedikit lebih tinggi dibandingkan penggunaan AI untuk homework. Selain itu, penggunaan AI untuk homework menunjukkan variasi kenaikan nilai yang lebih kecil, yang mengindikasikan hasil yang lebih konsisten di antara siswa

```python
df_ai['delta_nilai'] = df_ai['nilai_sesudah'] - df_ai['nilai_sebelum']
```

```python
df_ai[['nilai_sebelum','nilai_sesudah','delta_nilai']].describe()
```

Berdasarkan analisis deskriptif terhadap 40 siswa yang menggunakan AI, rata-rata nilai akademik meningkat dari 65,33 sebelum penggunaan AI menjadi 75,15 setelah penggunaan AI. Rata-rata peningkatan nilai sebesar 9,83 poin dengan seluruh siswa mengalami kenaikan nilai, yang mengindikasikan adanya kecenderungan positif penggunaan AI terhadap kinerja akademik

```python
df_ai[['rerata_jam_perhari', 'waktu_layar_harian', 'delta_nilai']].describe()
```

Berdasarkan analisis deskriptif, rata-rata siswa menggunakan AI selama 2,83 jam per hari dengan waktu layar harian sebesar 4,48 jam. Kenaikan nilai akademik tidak menunjukkan peningkatan yang sebanding dengan durasi penggunaan AI maupun waktu layar harian. Hal ini mengindikasikan bahwa peningkatan kinerja akademik lebih dipengaruhi oleh cara dan tujuan penggunaan AI dibandingkan lamanya waktu penggunaan
